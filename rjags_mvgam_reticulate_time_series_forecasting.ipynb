## RJAGS

```r
prepare_data_rjags <- function(data, test_steps = 24, K = 2, season_length = 12) {
  N <- nrow(data)
  train_data <- data[1:(N - test_steps), ]
  test_data  <- data[(N - test_steps + 1):N, ]

  y_train <- train_data$Passengers
  n_train <- length(y_train)

  # Build design matrices for in-sample (train)
  cosstuff_train <- matrix(NA, nrow = n_train, ncol = K)
  sinstuff_train <- matrix(NA, nrow = n_train, ncol = K)
  for (t in 1:n_train) {
    for (k in 1:K) {
      cosstuff_train[t, k] <- cos(2 * pi * k * t / season_length)
      sinstuff_train[t, k] <- sin(2 * pi * k * t / season_length)
    }
  }

  # Build design matrices for forecast horizon
  cosstuff_fore <- matrix(NA, nrow = test_steps, ncol = K)
  sinstuff_fore <- matrix(NA, nrow = test_steps, ncol = K)
  for (t in 1:test_steps) {
    for (k in 1:K) {
      # Time index is (n_train + t) to continue seamlessly
      cosstuff_fore[t, k] <- cos(2 * pi * k * (n_train + t) / season_length)
      sinstuff_fore[t, k] <- sin(2 * pi * k * (n_train + t) / season_length)
    }
  }

  # JAGS data list
  rjags_data <- list(
    y = y_train,
    N = n_train,
    n_forecast = test_steps,
    K = K,
    cosstuff_train = cosstuff_train,
    sinstuff_train = sinstuff_train,
    cosstuff_fore  = cosstuff_fore,
    sinstuff_fore  = sinstuff_fore
  )

  return(list(
    rjags_data = rjags_data,
    train_data = train_data,
    test_data  = test_data
  ))
}
```

```r
# Prepare the data
rjags_data_all <- prepare_data_rjags(
    airpassengers_data,
    test_steps = 24,
    K = 2,
    season_length = 12
)

# Extract train/test dataframes
train_data <- rjags_data_all$train_data
test_data  <- rjags_data_all$test_data
```

```r
prepare_data_mvgam <- function(df, test_steps = 24) {
  df_mod <- df %>%
    mutate(season = as.numeric(format(Month, "%m")),
           series = as.factor("AirPassengers"),
           time = 1:n())  # Create a complete time index
  N <- nrow(df_mod)
  train_data <- df_mod[1:(N - test_steps), ]
  test_data  <- df_mod[(N - test_steps + 1):N, ]

  return(list(
    train_data = train_data,
    test_data = test_data,
    test_idx = (N - test_steps + 1):N
  ))
}
```

```r
prepare_data_nf <- function(df, test_steps = 24) {
  # Convert to neuralforecast structure: (unique_id, ds, y)
  panel_df <- df %>%
    dplyr::mutate(
      ds = dplyr::row_number(),
      y  = Passengers,
      unique_id = "AirPassengers"
    ) %>%
    dplyr::select(ds, y, unique_id)

  # Train/Test split
  N <- nrow(panel_df)
  train_size <- N - test_steps

  train_df <- panel_df[1:train_size, ]
  test_df  <- panel_df[(train_size + 1):N, ]

  return(list(
    train_df = train_df,
    test_df = test_df
  ))
}
```

## RJAGS

```r
model_string <- "
model {
  pi <- 3.141592653589793

  # ------------------------------
  # 1) In-sample model
  # ------------------------------
  for (t in 1:N) {
    # mu_fitted[t] = mu[t] + sum(a[k]*cosstuff_train[t,k] + b[k]*sinstuff_train[t,k])
    mu_fitted[t] <- mu[t]
                    + inprod(a[], cosstuff_train[t,])
                    + inprod(b[], sinstuff_train[t,])

    y[t] ~ dnorm(mu_fitted[t], tau.obs)
  }

  # ------------------------------
  # 2) Latent states (local linear trend)
  # ------------------------------
  mu[1]    ~ dnorm(0, 1.0E-6)
  delta[1] ~ dnorm(0, 1.0E-6)
  for (t in 2:N) {
    mu[t]    ~ dnorm(mu[t-1] + delta[t-1], tau.mu)
    delta[t] ~ dnorm(delta[t-1], tau.delta)
  }

  # ------------------------------
  # 3) Fourier coefficients
  # ------------------------------
  for (k in 1:K) {
    a[k] ~ dnorm(0, 1.0E-6)
    b[k] ~ dnorm(0, 1.0E-6)
  }

  # ------------------------------
  # 4) Forecast block
  # ------------------------------
  mu_forecast[1]    <- mu[N]
  delta_forecast[1] <- delta[N]

  for (t in 1:n_forecast) {
    mu_forecast[t+1]    ~ dnorm(mu_forecast[t] + delta_forecast[t], tau.mu)
    delta_forecast[t+1] ~ dnorm(delta_forecast[t], tau.delta)

    # Sum of Fourier terms for forecast step t
    fourier_forecast[t] <- inprod(a[], cosstuff_fore[t,])
                           + inprod(b[], sinstuff_fore[t,])

    y_pred[t] ~ dnorm(mu_forecast[t+1] + fourier_forecast[t], tau.obs)
  }

  # ------------------------------
  # 5) Priors for precision
  # ------------------------------
  tau.obs   ~ dgamma(0.001, 0.001)
  tau.mu    ~ dgamma(0.001, 0.001)
  tau.delta ~ dgamma(0.001, 0.001)
}
"
```

```r
fit_rjags_model <- function(model_string, rjags_data,
                           n.iter = 5000,
                           n.burnin = 5000,
                           n.chains = 4,
                           n.thin = 1) {

  # Initialize model
  jags_model <- jags.model(
    textConnection(model_string),
    data = rjags_data,
    n.chains = n.chains,
    quiet = TRUE
  )

  update(jags_model, n.iter = n.burnin, progress.bar = "none")

  # Sample
  samples <- coda.samples(
    model = jags_model,
    variable.names = c("mu", "delta", "a", "b",
                       "mu_fitted", "y_pred",
                       "mu_forecast", "delta_forecast"),
    n.iter = n.iter,
    thin   = n.thin
  )
  return(samples)
}
```

```r
# Fit the model
set.seed(123)
fit_rjags <- fit_rjags_model(
  model_string,
  rjags_data_all$rjags_data,
  n.iter = 5000,
  n.burnin = 5000,
  n.chains = 4
)
```

# Model Forecasting

## RJAGS

```r
forecast_rjags <- function(fit,
                           test_data,
                           train_data = NULL,
                           include_train = FALSE) {

  # 1) Out-of-sample forecasts
  y_pred_samples <- do.call(rbind, lapply(fit, function(chain) {
    chain[, grep("^y_pred\\[", colnames(chain))]
  }))

  # Compute mean and intervals
  y_pred_mean <- apply(y_pred_samples, 2, mean)
  lower_y_pred <- apply(y_pred_samples, 2, quantile, probs = 0.025)
  upper_y_pred <- apply(y_pred_samples, 2, quantile, probs = 0.975)

  # Construct a dataframe for the test set
  forecast_test <- data.frame(
    Month = test_data$Month,
    Observed = test_data$Passengers,
    Predicted = y_pred_mean,
    Lower = lower_y_pred,
    Upper = upper_y_pred,
    row.names = NULL
  )

  if (!include_train) {
    # Return only out-of-sample (test) predictions
    return(forecast_test)
  } else {
    if (is.null(train_data)) {
      stop("When include_train = TRUE, 'train_data' must be provided.")
    }

    # 2) In-sample fitted values (mu_fitted = mu[t] + fourier[t])
    mu_fitted_samples <- do.call(rbind, lapply(fit, function(chain) {
      chain[, grep("^mu_fitted\\[", colnames(chain))]
    }))

    # Compute mean and intervals
    mu_fitted_mean <- apply(mu_fitted_samples, 2, mean)
    lower_mu_fitted <- apply(mu_fitted_samples, 2, quantile, probs = 0.025)
    upper_mu_fitted <- apply(mu_fitted_samples, 2, quantile, probs = 0.975)

    # Combine in-sample + out-of-sample into one dataframe
    all_months <- c(train_data$Month, test_data$Month)
    all_observed <- c(train_data$Passengers, test_data$Passengers)

    # Fitted + forecast means
    all_predicted <- c(mu_fitted_mean, y_pred_mean)
    all_lower <- c(lower_mu_fitted, lower_y_pred)
    all_upper <- c(upper_mu_fitted, upper_y_pred)

    full_df <- data.frame(
      Month = all_months,
      Observed = all_observed,
      Predicted = all_predicted,
      Lower = all_lower,
      Upper = all_upper,
      row.names = NULL
    )

    return(full_df)
  }
}
```

## MVGAM

```r
forecast_mvgam <- function(fit, test_idx, dates, observed, n_samples = 2000) {
  # Run the forecast
  h <- length(test_idx)
  fc <- forecast(fit, h = h, n_samples = n_samples)

  # Extract the samples
  samples <- fc$forecasts$AirPassengers

  # Compute mean and intervals
  pred_means <- apply(samples, 2, mean)
  pred_lower <- apply(samples, 2, quantile, probs = 0.025)
  pred_upper <- apply(samples, 2, quantile, probs = 0.975)

  # Build the forecast dataframe
  forecast_df <- data.frame(
    Month = dates[test_idx],
    Observed = observed[test_idx],
    Predicted = pred_means,
    Lower = pred_lower,
    Upper = pred_upper,
    row.names = NULL
  )

  return(forecast_df)
}
```

## NeuralForecast

```r
forecast_nf <- function(setup_nf, test_df, level = 95, first_forecast_date = "1959-01-01", freq = "month") {
  # Obtain forecasts using the fitted RMOK model
  forecast_results <- setup_nf$predict(futr_df=test_df, level=list(as.integer(level)))
  forecast_df <- reticulate::py_to_r(forecast_results)
  forecast_df <- as.data.frame(forecast_df)

  # Remove 'unique_id' column
  if ("unique_id" %in% names(forecast_df)) {
    forecast_df$unique_id <- NULL
  }

  # Rename forecast columns
  if ("RMoK" %in% names(forecast_df)) {
    names(forecast_df)[names(forecast_df) == "RMoK"] <- "Predicted"
  }
  if ("RMoK-lo-95" %in% names(forecast_df)) {
    names(forecast_df)[names(forecast_df) == "RMoK-lo-95"] <- "Lower"
  }
  if ("RMoK-hi-95" %in% names(forecast_df)) {
    names(forecast_df)[names(forecast_df) == "RMoK-hi-95"] <- "Upper"
  }

  # Generate forecast dates based on the first_forecast_date
  horizon <- nrow(test_df)
  forecast_dates <- seq(from = as.Date(first_forecast_date),
                        by = freq,
                        length.out = horizon)
  forecast_df$Month <- forecast_dates

  # Process the test_df to extract observed values
  test_df_r <- reticulate::py_to_r(test_df)
  test_df_r <- as.data.frame(test_df_r)
  test_df_r$Month <- forecast_dates

  # Rename the 'y' column to 'Observed'
  if ("y" %in% names(test_df_r)) {
    test_df_r <- dplyr::rename(test_df_r, Observed = y)
  }

  # Merge forecast with observed values on the Month column
  final_df <- merge(forecast_df, test_df_r[, c("Month", "Observed")],
                    by = "Month", all.x = TRUE)

  # Re-order the columns
  final_df <- final_df[, c("Month", "Observed", "Predicted", "Lower", "Upper")]

  return(final_df)
}
```

```r
# Forecast
forecast_df3 <- forecast_nf(
    setup_nf,
    test_df,
    level = 95,
    first_forecast_date = "1959-01-01",
    freq = "month"
)
print("Forecasts for test data:")
head(forecast_df3)
```

# Metrics Evaluation

```r
compute_metrics <- function(forecast_df) {
  eps <- .Machine$double.eps  # a very small constant to avoid division by zero
  rmse  <- sqrt(mean((forecast_df$Predicted - forecast_df$Observed)^2, na.rm = TRUE))
  mae   <- mean(abs(forecast_df$Predicted - forecast_df$Observed), na.rm = TRUE)
  mape  <- mean(abs((forecast_df$Predicted - forecast_df$Observed) / (forecast_df$Observed + eps)), na.rm = TRUE) * 100
  smape <- mean(2 * abs(forecast_df$Predicted - forecast_df$Observed) /
                  (abs(forecast_df$Predicted) + abs(forecast_df$Observed) + eps), na.rm = TRUE) * 100

  metrics_df <- data.frame(
    RMSE = rmse,
    MAE = mae,
    MAPE = mape,
    SMAPE = smape
  )
  return(metrics_df)
}
```

```r
# Compute metrics
metrics_df1 <- compute_metrics(forecast_df1)
metrics_df2 <- compute_metrics(forecast_df2)
metrics_df3 <- compute_metrics(forecast_df3)

final_metrics_df <- rbind(
  data.frame(Model = "RJAGS", metrics_df1),
  data.frame(Model = "MVGAM", metrics_df2),
  data.frame(Model = "NeuralForecast", metrics_df3)
)

final_metrics_df
```

```r
plot_all_forecasts <- function(forecast_list, title) {
  # forecast_list: a named list of forecast dataframes
  # Each forecast data frame must contain:
  #   Month, Observed, Predicted, Lower, Upper

  # Add a 'Model' column to each dataframe based on the list name
  forecast_list <- lapply(names(forecast_list), function(model_name) {
    df <- forecast_list[[model_name]]
    df$Model <- model_name
    df
  })

  # Combine all forecasts into one dataframe
  combined_df <- do.call(rbind, forecast_list)

  # Since the observed values should be the same across forecasts,
  # we extract a distinct observed series
  observed_df <- combined_df %>%
    select(Month, Observed) %>%
    distinct()

  # Create the plot
  p <- ggplot() +
    # Plot predicted lines for each model
    geom_line(data = combined_df,
              aes(x = Month, y = Predicted, color = Model),
              linewidth = 1.2) +
    # Plot the observed values as a black dashed line
    geom_line(data = observed_df,
              aes(x = Month, y = Observed),
              color = "black", linewidth = 1.2, linetype = "dashed") +
    labs(title = title, x = "Month", y = "Passengers (in thousands)") +
    theme_minimal() +
    guides(color = guide_legend(reverse = TRUE))

  p <- p +
    scale_color_manual(name = "Model",
                       values = c("RJAGS" = "blue",
                                  "MVGAM" = "green",
                                  "NeuralForecast" = "red"))

  return(p)
}
```

# Ensemble Forecast

```r
# Define a function to compute sMAPE for a given weight
smape_error <- function(w, f1, f2, obs, epsilon = 1e-8) {
  # Compute the ensemble forecast as a weighted average
  pred <- w * f1 + (1 - w) * f2

  # Calculate sMAPE
  errors <- 2 * abs(pred - obs) / (abs(pred) + abs(obs) + epsilon)

  return(mean(errors))
}
```

```r
# Extract forecasts and actual values
f1 <- forecast_df1$Predicted
f2 <- forecast_df3$Predicted
obs <- forecast_df1$Observed

# Optimize the weight in the interval [0, 1]
opt_result <- optimize(smape_error,
                       interval = c(0, 1),
                       f1 = f1,
                       f2 = f2,
                       obs = obs)

optimal_weight <- opt_result$minimum
cat("Optimal weight for forecast_df1:", optimal_weight, "\n")
cat("Optimal weight for forecast_df3:", 1 - optimal_weight, "\n")
```

```r
# Compute the weighted ensemble forecast using the optimal weight
ensemble_forecast <- forecast_df1

ensemble_forecast$Predicted <- optimal_weight * forecast_df1$Predicted +
                               (1 - optimal_weight) * forecast_df3$Predicted

ensemble_forecast$Lower <- optimal_weight * forecast_df1$Lower +
                           (1 - optimal_weight) * forecast_df3$Lower

ensemble_forecast$Upper <- optimal_weight * forecast_df1$Upper +
                           (1 - optimal_weight) * forecast_df3$Upper

head(ensemble_forecast)
```

```r
plot_all_forecasts_updated <- function(forecast_list,
                                       title,
                                       palette = NULL,
                                       include_history = FALSE,
                                       historical_data = NULL,
                                       hist_date_col = "Month",
                                       hist_target_col = "Observed") {

  # Add a 'Model' column based on the list names
  forecast_list <- lapply(names(forecast_list), function(model_name) {
    df <- forecast_list[[model_name]]
    df$Model <- model_name
    df
  })

  # Combine forecast dataframes into one dataframe
  combined_df <- do.call(rbind, forecast_list)

  # Extract the observed values
  observed_df <- combined_df %>%
    select(Month, Observed) %>%
    distinct()

  # Determine the historical data to plot
  if (include_history) {
    if (is.null(historical_data)) {
      stop("Historical data must be provided when include_history = TRUE.")
    }
    hist_data <- historical_data
  } else {
    hist_data <- combined_df %>%
      select(Month, Observed) %>%
      distinct()
  }

  # Determine unique model names
  models <- unique(combined_df$Model)

  # If no palette is provided, generate one dynamically
  if (is.null(palette)) {
    palette <- hue_pal()(length(models))
    names(palette) <- models
  }

  # Build the plot
  p <- ggplot() +
    # Plot predicted lines for each model
    geom_line(data = combined_df,
              aes(x = Month, y = Predicted, color = Model),
              linewidth = 1.2) +
    # Plot the observed values as a black dashed line
    geom_line(data = observed_df,
              aes(x = Month, y = Observed),
              color = "black", linewidth = 1.2, linetype = "dashed")

  # Plot the historical data as a dashed line
  if (include_history) {
    p <- p + geom_line(data = hist_data,
                       aes(x = .data[[hist_date_col]], y = .data[[hist_target_col]]),
                       color = "black", linewidth = 1.2, linetype = "dashed")
  } else {
    p <- p + geom_line(data = hist_data,
                       aes(x = Month, y = Observed),
                       color = "black", linewidth = 1.2, linetype = "dashed")
  }

  # Finalize plot appearance
  p <- p +
    labs(title = title, x = "Month", y = "Passengers (in thousands)") +
    theme_minimal() +
    guides(color = guide_legend(reverse = TRUE)) +
    scale_color_manual(name = "Model", values = palette)

  return(p)
}
```

```r
split_data <- function(data, test_steps = 24) {
  N <- nrow(data)
  train_data <- data[1:(N - test_steps), ]
  test_data  <- data[(N - test_steps + 1):N, ]

  return(list(
    train_data = train_data,
    test_data  = test_data
  ))
}
```

```r
# Prepare the data
history_df <- split_data(
    airpassengers_data2,
    test_steps = 24
)

# Extract train/test dataframes
train_hist <- history_df$train_data
test_hist  <- history_df$test_data
```
